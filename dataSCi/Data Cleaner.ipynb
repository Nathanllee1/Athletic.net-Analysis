```python
import pandas as pd
```

```python
df = pd.read_csv("database.csv")
```

```python
df
```

```python
def replaceErrors(result):
    strResult = str(result)
    return(strResult.replace("PR", "").replace("SR", "").replace("h", "0").replace("c", ""))
```

```python
def removeParen(result):
    
    return re.sub("[\(\[].*?[\)\]]", "", result)
```

```python
def removeParen2(result):
    if "(" in result and ")" not in result:
        return result[:-4]
    else:
        return result
```

```python
import re
```

```python
formattedDF = df["result"].apply(replaceErrors).apply(removeParen).apply(removeParen2)
```

```python
df['formattedResults'] = formattedDF
```

```python
df = df.drop(columns=['result'])
df = df.drop(columns=['Unnamed: 0'])
```

```python
df
```

```python
df = df[df.formattedResults != "ND"]
```

```python
df = df[df.formattedResults != "NT"]
```

```python
def removeCharacters(result):
    return re.sub("[a-zA-Z]", "", result)
```

```python
noAlpha = df['formattedResults'].apply(removeCharacters)
```

```python
df['formattedResults'] = noAlpha
```

```python
df
```

```python
def percentileConverter(result):
    
    try:
        result = str(result)
        
        if ':' in result:
            splitResult = re.split(":", result)
            
            if len(splitResult) == 1:
                return float(splitResult)
            if len(splitResult) == 2:
                
                return float(splitResult[0]) * 60 + float(splitResult[1])
            if len(splitResult) == 3:
                return float(splitResult[0]) * 60 + float(splitResult[1]) * 60 + float(splitResult[2])
        if "'" in result:
            splitDistance = re.split("'", result)
            if len(splitDistance) == 1:
                return float(splitDistance)
            if len(splitDistance) == 2:
                return float(splitDistance[0]) * 12 + float(splitDistance[1])
        else:
            return float(result)
    except ValueError:
        #print('uhoh')
        return "Error"
```

```python
df['splitTime'] = df['formattedResults'].apply(percentileConverter)
```
